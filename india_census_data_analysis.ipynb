## Visual Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as px
```

### Visualize the population distribution across different states.

```python
state_population = df.groupby('State_name')['Population'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
state_population.plot(kind='bar', color='skyblue')
plt.title('Population Distribution by State (India Census 2011)')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

### Gender Distribution across districts

```python
plt.figure(figsize=(10, 6))
sb.histplot(data=df, x='Male', color='blue', label='Male', kde=True)
sb.histplot(data=df, x='Female', color='pink', label='Female', kde=True)
plt.title('Gender Distribution (India Census 2011)')
plt.xlabel('Population')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Literacy Rate Distribution across districts

```python
fig = px.histogram(df, x='Literacy_Rate', nbins=30, title='Literacy Rate Distribution (India Census 2011)')
fig.update_layout(xaxis_title='Literacy Rate (%)', yaxis_title='Count')
fig.update_traces(marker=dict(line=dict(color='black', width=1)))
fig.show()
```

### Distribution of Workers across states

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Workers'], bins=20, color='orange', alpha=0.7, edgecolor='black')
plt.title('Distribution of Workers (India Census 2011)')
plt.xlabel('Number of Workers')
plt.ylabel('Frequency')
plt.show()
```

### Population vs. Literacy Rate

```python
plt.figure(figsize=(12, 8))

for state, color in zip(df['State_name'].unique(), sb.color_palette('tab10', n_colors=len(df['State_name'].unique()))):
    state_data = df[df['State_name'] == state]
    sb.scatterplot(data=state_data, x='Population', y='Literacy_Rate', color=color, alpha=0.7, label=state)

plt.title('Population vs. Literacy Rate (India Census 2011)')
plt.xlabel('Population')
plt.ylabel('Literacy Rate (%)')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
